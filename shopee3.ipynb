```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv')
df.head(1)
```

##### Vẽ biểu đồ tần số số lượng shop gia nhập theo các năm.

```python
sns.countplot(y ='join_year', hue = "join_year", data = df)
plt.grid()
```

##### Vẽ biểu đồ xu hướng thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt.

```python
sns.lmplot(x="rating_good", y="response_rate", data=df)
plt.title('Xu hướng mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt')
```

##### Vẽ biểu đồ xu hướng thể hiện mối quan hệ giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu.

```python
df['response_time'] = pd.to_datetime(df['response_time'].str.strip())
```

```python
sns.regplot(x = 'rating_bad', y = df['response_time'].dt.second, data = df)
plt.grid()
plt.title('Xu hướng mối quan hệ giữa rating_bad và response_time');
```

##### Vẽ biểu đồ thể hiện phân bố của điểm đánh giá trung bình.

```python
sns.violinplot(y="rating_normal", data=df);
plt.grid()
plt.title('Phân bố điểm đánh giá trung bình')
```

##### Vẽ biểu đồ tần số của cửa hàng chính thức và không chính thức.

```python
sns.countplot(x ='is_official_shop', data=df)
```
